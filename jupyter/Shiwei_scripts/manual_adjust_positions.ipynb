```python
import numpy as np
import os,glob,sys
```

```python
old_manual_positions = [[-4823.30,-1341.10]
                        ,]

new_manual_positions = [[-4821.30, -1342.10],
                        ]

shifts = np.array(new_manual_positions) - np.array(old_manual_positions)
print(shifts)
# calculate mean
manual_shift = np.round(np.mean(shifts, axis=0), 2)
print(manual_shift)
```

```python
data_folder = r'D:\Shiwei\20240119-RPE1_CTP14_genome_Tn5_exp2'
old_positions = np.loadtxt(os.path.join(data_folder, 'positions_all.txt'), delimiter=',')
```

```python
print(old_positions)
```

```python
adjusted_positions = old_positions + manual_shift
```

```python
adj_save_filename = os.path.join(data_folder, 'adjusted_positions_all.txt')
print(adj_save_filename)
np.savetxt(adj_save_filename, adjusted_positions, fmt='%.2f', delimiter=',')
```
