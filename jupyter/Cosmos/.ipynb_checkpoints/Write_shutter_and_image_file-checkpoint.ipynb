```python
# write shutter file

counter = 0

for i in range(10):
    
    for channel in [0, 1, 3, 4]:
        print(r'<event>')
        print('\t<channel>'+str(channel)+'</channel>')
        print('\t<power>1.0</power>')
        print('\t<on>'+str(counter)+'</on>')
        print('\t<off>'+str(counter+1)+'</off>')
        print('\t<color>255,255,255</color>')
        counter += 1
        print('</event>')
    print()
```
